```python
import pandas as pd #import library pandas

dataset = pd.read_csv("content/train.csv") #data loading 

dataset.head() #menampilkan isi dataset
```

```python
dataset.info()
```

```python
#.describe digunakan untuk menampilkan statistika deskriptif dari dataset tsb 

dataset.describe(include="all")
```

```python
missing_values = dataset.isnull().sum()
missing_values[missing_values > 0]
```

```python
less = missing_values[missing_values < 1000].index
over = missing_values[missing_values >= 1000].index
```

```python
numeric_features = dataset[less].select_dtypes(include=['number']).columns
dataset[numeric_features] = dataset[numeric_features].fillna(dataset[numeric_features].median())
```

```python
kategorical_features = dataset[less].select_dtypes(include=['object']).columns
 
for column in kategorical_features:
    dataset[column] = dataset[column].fillna(dataset[column].mode()[0])
```

```python
df = dataset.drop(columns=over)
```

```python
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


for feature in numeric_features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[feature])
    plt.title(f'Box Plot of {feature}')
    plt.show()
```
